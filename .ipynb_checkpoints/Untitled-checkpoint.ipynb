# matplotlib

```python
import pandas as pd
test = pd.read_csv('test.csv')
test.head()
#test.columns
```

```python
import matplotlib.pyplot as plt
plt.plot(test['PassengerId'],test['Fare'])
#横坐标斜45度
plt.xticks(rotation=45)
#x轴y轴标签
plt.xlabel('passergerId')
plt.ylabel('fare')
#图的名称
plt.title('test')
plt.show()
```

```python
#设置一个子图的空间
fig = plt.figure()
#4个子图，排成2行2列，ax1为第一个子图，ax2为第二个
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax4 = fig.add_subplot(2,2,4)
plt.show()
```

```python
import numpy as np
#figsize设置图形的长和高
fig = plt.figure(figsize=(6,3))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))
plt.show()
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(test['PassengerId'][0:10],test['Fare'][0:10],c='red',label='t1')
plt.plot(test['PassengerId'][10:20],test['Fare'][10:20],c='blue',label='t2')
plt.legend(loc='best')
plt.show()
#打印plt.legend的帮助文档，loc参数可以设置成'upper right','upper left'等
#print(help(plt.legend))
```
